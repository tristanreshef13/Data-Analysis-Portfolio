```python
#שאלה 2 סעיף 2
df_no_vegen=df[df['pizza_type'] != 'Veggie']
```

```python
#שאלה 2 סעיף 3
# הסרת טורים שלא מוסיפים מידע להשוואה בין סוגי הפיצות
df_clean= df_no_vegen.drop(columns=['Unnamed: 0', 'order_details_id', 'order_time', 'pizza_name', 'is_it_pizza'])
```



**Unnamed: 0** – This is a redundant column as it only counts the rows and does not contain any useful information.  
**order_details_id** – A unique identifier for each order. It has no comparative value between different types of pizzas.  
**order_time** – This column indicates the order time, but for the purpose of comparing pizza types (and not order times), it is irrelevant.  
**pizza_name** – The pizza names might be unique to each pizza type, and therefore they do not add any additional information beyond the pizza_type column and the ingredians column.  
**is_it_pizza** – All rows contain the value "Yes," so this column has no value when comparing different pizza types.

```python
#שאלה 2 סעיף 4
# חישוב ממוצע, סטיית תקן ומספר התצפיות לפי סוג פיצה וכמות
price_stats = df_clean.groupby(['pizza_type', 'quantity']).agg(
    avg_total_price=('total_price', 'mean'),
    std_total_price=('total_price', 'std'),
    count=('total_price', 'count')
).reset_index()

# הדפסת התוצאות
print("Statistics (mean, std, count) by pizza type and quantity:")
print(price_stats)

```

#התגלו לנו סטיות תקן מאוד קיצוניות עבור סקטור האוכל,נצטרך לבחון את הדאטה יותר לעומק ולנקות אותו

```python
df2=df_clean[df_clean['total_price'] == df_clean['quantity'] * df_clean['unit_price']]
df2= df2[(df2['total_price'] >0)]
df2.describe()
#אותו תהליך ניקוי שעשיתי בשאלה הבאה
```

```python
# חישוב ממוצע, סטיית תקן, חציון ומספר תצפיות לפי סוג פיצה וכמות
price_stats =df2.groupby(['pizza_type', 'quantity']).agg(
    avg_unit_price=('unit_price', 'mean'),
    std_unit_price=('unit_price', 'std'),
    median_unit_price=('unit_price', 'median'),
    count=('unit_price', 'count')
).reset_index()

# חישוב סך כל התצפיות
total_count = price_stats['count'].sum()

# הוספת עמודת אחוז מכלל התצפיות
price_stats['percent_of_total'] = (price_stats['count'] / total_count) * 100

print("Statistics (mean, std, median, count, percent) by pizza type and quantity:")
print(price_stats)
```

We start by checking how many pizzas the company produces each day to get a clearer picture. For this, we use `zip` – a function that combines two (or more) lists into pairs.
```python
df5 = df_without_mistakes_question3
#נשתמש בקופי-פונקציה שיוצרת דאטה בייס חדש שלא משפיע על
#הדאטה הקודם כדי למנוע טעויות שישנו לנו דברים בדאטה המקורי
df_monday = df5[df5['day_of_week'] == 'Monday'].copy()
df_tuesday = df5[df5['day_of_week'] == 'Tuesday'].copy()
df_wednesday = df5[df5['day_of_week'] == 'Wednesday'].copy()
df_thursday = df5[df5['day_of_week'] == 'Thursday'].copy()
df_friday = df5[df5['day_of_week'] == 'Friday'].copy()
df_saturday = df5[df5['day_of_week'] == 'Saturday'].copy()
df_sunday = df5[df5['day_of_week'] == 'Sunday'].copy()

def print_time_of_day_and_quantity():
    # List of DataFrames for each day of the week
    days_data = [df_monday, df_tuesday, df_wednesday, df_thursday, df_friday, df_saturday, df_sunday]
    days_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


    for day_df, day_name in zip(days_data, days_names):
        print(f"Total quantity of pizzas for each time of day on {day_name}:")

        # Multiply 'quantity' by 1 to get the total number of pizzas ordered for each time of day
        total_pizzas = day_df.groupby('time_of_day')['quantity'].sum()

        # Calculate the total pizzas (morning + afternoon + evening)
        total_pizzas_all_times = total_pizzas.sum()  # Sum of morning, afternoon, and evening

        print(total_pizzas)
        print(f"Total pizzas for {day_name} (morning + afternoon + evening): {total_pizzas_all_times}")
        print("\n" + "-"*40)

print_time_of_day_and_quantity()
```

```python
from colorama import Fore, Style
total_pizzas_week = df5['quantity'].sum()
def salary_budget_distribution():
    # List of DataFrames for each day of the week
    days_data = [df_monday, df_tuesday, df_wednesday, df_thursday, df_friday, df_saturday, df_sunday]
    days_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

    # Define rainbow colors
    rainbow_colors = [Fore.RED, Fore.YELLOW, Fore.GREEN, Fore.CYAN, Fore.BLUE, Fore.MAGENTA]

    # Iterate through each DataFrame
    for day_df, day_name in zip(days_data, days_names):
        print(f"{rainbow_colors[0]}Total quantity of pizzas for each time of day on {day_name}:{Style.RESET_ALL}")

        # Calculate the total number of pizzas ordered for each time of day (morning, afternoon, evening)
        total_pizzas = day_df.groupby('time_of_day')['quantity'].sum()

        # Calculate the total pizzas (morning + afternoon + evening)
        total_pizzas_daily = total_pizzas.sum()

        # Calculate the percentage of each time of day relative to the total pizzas for the week
        percentage_pizzas = total_pizzas_daily / total_pizzas_week * 100

        # Calculate the percentage of total pizzas for the day relative to the weekly total
        percentage_day = total_pizzas_daily / total_pizzas_week * 100

        # Print the total pizzas for each time of day and the percentage of the weekly total
        print(f"{rainbow_colors[1]}{total_pizzas}{Style.RESET_ALL}")
        print(f"{rainbow_colors[2]}Total pizzas for {day_name} (morning + afternoon + evening): {total_pizzas_daily}{Style.RESET_ALL}")
        print(f"{rainbow_colors[3]}The daily pizza units divided by the weekly pizza amount would give us the percentage of budget we want to allocate to that day/amount of time.{Style.RESET_ALL}")
        print(f"{rainbow_colors[4]}Percentage of total budget that should go to salaries on {day_name}:{Style.RESET_ALL}")
        print(f"{rainbow_colors[5]}{percentage_pizzas}{Style.RESET_ALL}")
        print(f"{rainbow_colors[0]}Percentage of total pizzas for the week on {day_name}: {percentage_day:.2f}%{Style.RESET_ALL}")
        print("\n" + "-" * 40)

salary_budget_distribution()
```
